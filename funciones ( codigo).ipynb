```python
# funciones.py

import random

def pedir_coordenadas():
    """Pide coordenadas al usuario (validación simple)"""
    while True:
        try:
            x = int(input("Fila (0-9): "))
            y = int(input("Columna (0-9): "))
            if 0 <= x <= 9 and 0 <= y <= 9:
                return x, y
            else:
                print("Coordenadas fuera del rango.")
        except ValueError:
            print("Introduce números válidos.")

def disparo_maquina():
    """Genera coordenadas aleatorias para la máquina"""
    x = random.randint(0, 9)
    y = random.randint(0, 9)
    print(f"La máquina dispara a ({x}, {y})")
    return x, y
```
